## Na zaklade analyzy dat jsem vyrobili tuto uber-funkci ktera provede vsechny diskutovane ukony najednou

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import warnings
from sklearn.metrics import classification_report

warnings.filterwarnings('ignore') # haha. I figured out solution for all warings.

def preprocess_dataframe(path_to_csv):
    df = pd.read_csv(path_to_csv)
    le = LabelEncoder()
    
    #########
    # 0. id #
    #########
    df = df.drop(columns=['ID'])
    
    ############
    # 1. class #
    ############
    # df.pclass = le.fit_transform(df.pclass)
    pclass_map = {
    1 : 2,
    2 : 1,
    3 : 0,
    }
    df.pclass_map = df.pclass.replace(pclass_map)
    
    ###########
    # 2. name #
    ###########
    # nobles
    df['noble'] = 0
    df['noble'][df.name.str.contains('Countess\.')] = 4
    df['noble'][df.name.str.contains('Lady\.')] = 3
    df['noble'][df.name.str.contains('Don\.')] = 2
    df['noble'][df.name.str.contains('Dona\.')] = 2
    df['noble'][df.name.str.contains('Jonkheer\.')] = 1
    
    # military
    df['military'] = 0
    df['military'][df.name.str.contains('Col\.')] = 3
    df['military'][df.name.str.contains('Major\.')] = 2
    df['military'][df.name.str.contains('Capt\.')] = 1
    
    # honorific
    df['honorific'] = np.NaN
    df['honorific'][df.name.str.contains('Mr\.')] = 'Mr'
    df['honorific'][df.name.str.contains('Master\.')] = 'Master'
    df['honorific'][df.name.str.contains('Ms\.')] = 'Ms'
    df['honorific'][df.name.str.contains('Miss\.')] = 'Miss'
    df['honorific'][df.name.str.contains('Mlle\.')] = 'Miss'
    df['honorific'][df.name.str.contains('Mrs\.')] = 'Mrs'
    df['honorific'][df.name.str.contains('Mme\.')] = 'Mrs'
    df['honorific'][df.name.str.contains('Rev\.')] = 'Rev'
    df['honorific'][df.name.str.contains('Dr\.')] = 'Dr'

    honors_onehot = pd.get_dummies(df['honorific'],prefix='h')
    df = pd.concat([df.drop(columns=['honorific']), honors_onehot], axis=1)
    
    if (path_to_csv == 'evaluation.csv'):
        df['h_Ms'] = 0
        df.h_Ms = df.h_Ms.astype(np.uint8)

    # drop name
    df = df.drop(columns=['name'])
    
    ##########
    # 3. sex #
    ##########
    df['sex'] = le.fit_transform(df['sex'])

    ##########
    # 4. age # + doplnit! (20%) [u testing take 20%]
    ########## 
    #  ... later ....

    #####################
    # 5. sipsp 6. parch #
    #####################
    df['whole_family'] = df.sibsp + df.parch

    #############
    # 7. ticket #
    #############
    df = df.drop(columns=['ticket'])

    ###########
    # 8. fare # +doplnit! (0.01%) [median from pclass] ; nic u testing
    ###########
    if (df['fare'].isnull().values.any() ) :
        df['fare'][df.fare.apply(np.isnan)] = df.loc[df.pclass == df.loc[df['fare'][df.fare.apply(np.isnan)].index].pclass.iloc[0] ].fare .median()
    
    bins = np.arange(1, 10) * 10 # naka fare byla i 500!
    df['fare_category'] = np.digitize(df.fare, bins, right=True)

    ############ 
    # 9. cabin # +doplnit! (80%, 20%fc) (deck-> Unknown, Cabin_Num -> 170)  ---> chtelo by to lepe !!!!!! (napr. podle fare / pclass)
    ############ u testing 81%
    # deck
    df.cabin = df.cabin.fillna('U') 
    df['deck'] = df.cabin.str.replace('[^a-zA-Z]+', '', regex=True).str[0]
    
    # deck label encoded
    deck_map = {
        'A' : 0,
        'B' : 1,
        'C' : 2,
        'D' : 3,
        'E' : 5,
        'F' : 6,
        'G' : 7,
        'U' : 8,
        'T' : 9
    }

    df.deck = df.deck.replace(deck_map)

    # cabin_num
    df['cabin_num'] = df.cabin.str.replace(' .*', '').str.replace('[^0-9]+', '', regex=True).str.replace('^$', '170', regex=True)
    df.cabin_num = df.cabin_num.fillna(170) # toto vypada podezrele, ale chci odlisit ty co neznam od tech co budou v kategori 0
    df.cabin_num = df.cabin_num.astype(np.int64)

    # cabin category
    bins = np.arange(1, 18) * 10
    df['cabin_category'] = np.digitize(df.cabin_num, bins, right=True)
    
    # samotnou cabin mazeme smazat
    df = df.drop(columns=['cabin'])

    ################ 
    # 10. embarked # +doplnit! (0.01%) (nejcastejnsi)
    ################ 1x u testing 
    df.embarked = df.embarked.fillna('S') 

    # OHC
    emb_onehot = pd.get_dummies(df['embarked'],prefix='em')
    df = pd.concat([df.drop(columns=['embarked']), emb_onehot], axis=1)
                                       
    #################
    # 11. home.dest # +doplnit! (40%, dest 55%) (???)
    ################# [55% u testing]
    df = df.drop(columns=['home.dest'])
    # okopiruju
    #df['dest'] = df['home.dest']
    # ty co nekonci na [a-z][a-z] budou null
    #df['dest'] = df['home.dest'].mask(df['home.dest'].str.contains('[A-Z][A-Z]$', regex=True) == False) 
    # ty co konci na [a-z][a-z] budou jen posledni dva znaky
    #df['dest'] = df['dest'].mask(df['dest'].str.contains('[A-Z][A-Z]$', regex=True) == True, other=df['dest'].str[-2:])

    return df

def preprocess_and_encode_age(filename):
    df = preprocess_dataframe(filename)
    five_categories = np.array([0.0, 16.133, 32.1, 48.067, 64.033, 81])
    df['age_group_enc'] = np.digitize(df.age, five_categories)
    df.loc[df.age.apply(np.isnan), 'age_group_enc'] = np.nan
    df = df.drop(columns= ['age'])
    return df
```

# k-fold Cross Validation && Iterace pres vsechny hyperparametry
### Zde jsem, abych predesel repetitivnosti kodu, pripravil funkce a metody, ktere udelalji 99% prace

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.naive_bayes import BernoulliNB


from sklearn.model_selection import train_test_split
from sklearn.model_selection import ParameterGrid
from sklearn.model_selection import KFold
import sklearn.metrics as metrics

# Cross validace
def cross_val(X, Y, folds, model, metric_func):
    
    kf = KFold(n_splits=folds, shuffle=False) 
    
     # e1 .. ek acc results
    test_acc_results = []
    train_acc_results = []

    # create K-fold cross validation
    for train_index, test_index in kf.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        Y_train, Y_test = Y.iloc[train_index], Y.iloc[test_index]
        # train model
        model.fit(X_train, Y_train)
        
        # measure model
        train_acc_results.append( metric_func(Y_train, model.predict(X_train)))
        test_acc_results.append( metric_func(Y_test, model.predict(X_test)))
    
    # return mean of metric
    return (np.array(train_acc_results).mean(), np.array(test_acc_results).mean())

# Class pro vraceni vysledku trenovani
class ModelResults:
    def __init__(self, cv_train_acc, cv_test_acc, holdout_acc):
        self.cv_train_acc = cv_train_acc
        self.cv_test_acc = cv_test_acc
        self.holdout_acc = holdout_acc

# Nic lepsiho me nenapadlo...
def get_model(model_name, params):
    if(model_name == 'decision_tree'):
        return DecisionTreeClassifier(**params)
    elif(model_name == 'random_forrest'):
        return RandomForestClassifier(**params)
    elif(model_name == 'ada_boost'):
        return AdaBoostClassifier(**params)
    elif(model_name == 'logistic_regression'):
        return LogisticRegression(**params)
    elif(model_name == 'knn'):
        return KNeighborsClassifier(**params)

# Lazeni hyperparametru (funkce mimo jine bere nazev modelu, metriku a take jestli vetsi hodnota metriky ==  lepsi )
def crosval_train(model_name, X, Y, X_holdout, Y_holdout, param_comb, num_of_folds, metric, greater):
    crossval_train_acc = []
    crossval_test_acc = []
    
    # for all hyper parametters
    for params in param_comb:   
        model = get_model(model_name, params)
        train_acc, test_acc = cross_val(X, Y,num_of_folds,model, metric)  
        crossval_test_acc.append(test_acc)
        crossval_train_acc.append(train_acc)
        
    # find best hyper parameters
    if greater:
        best_params = param_comb[np.argmax(crossval_test_acc)]
    else:
        best_params = param_comb[np.argmin(crossval_test_acc)]
       
    # train on best hyperparams and measure holdout acc
    model = get_model(model_name, best_params)
    model.fit(X,Y)
    holdout_acc = metric(Y_holdout, model.predict(X_holdout))

    return ModelResults(crossval_train_acc, crossval_test_acc, holdout_acc)

# nakonec jeste funkce pro snadne plotovani
def my_plot(result, plotname):
    fig = plt.figure(figsize=(20,6))
    plt.plot(result.cv_train_acc,'or-')
    plt.plot(result.cv_test_acc,'ob-')
    plt.plot([0,len(result.cv_test_acc)], [result.holdout_acc,result.holdout_acc],'g-')
    plt.xlabel('max_depth')
    plt.ylabel('accuracy')
    plt.legend(['train_acc_mean', 'test_acc_mean','houdlout_acc'])
    plt.title(plotname, fontsize=14, fontweight='bold')
    return fig
```

```python
# WARNING: this runs ~10 mins (2 core i7-6600U)
if(train_survived):
    num_fold = 10
    metric = metrics.accuracy_score
    greater = True

    results_to_plot = {}
    # DESICION TREE
    model_name = 'decision_tree'
    param_comb = ParameterGrid({'max_depth': range(1,30),  'criterion': ['entropy', 'gini']})

    result = crosval_train(model_name, X, Y, X_holdout, Y_holdout, param_comb, num_fold, metric, greater)
    results_to_plot[model_name] = result
    models_results_survived[model_name] = {'acc': result.holdout_acc, 'params' : param_comb[np.argmax(result.cv_test_acc)]}



    # RANDOM FORREST
    model_name = 'random_forrest'
    param_comb = ParameterGrid({'n_estimators': range(1,100,5), 'max_depth': range(1,10)})

    result = crosval_train(model_name, X, Y, X_holdout, Y_holdout, param_comb, num_fold, metric, greater)
    results_to_plot[model_name] = result
    models_results_survived[model_name] = {'acc': result.holdout_acc, 'params' : param_comb[np.argmax(result.cv_test_acc)]}



    # ADA BOOST
    model_name = 'ada_boost'
    param_comb = ParameterGrid({'n_estimators': range(1,100,5),'learning_rate': [0.01, 0.05, 0.1, 0.3, 0.5, 1]})

    result = crosval_train(model_name, X, Y, X_holdout, Y_holdout, param_comb, num_fold, metric, greater)
    results_to_plot[model_name] = result
    models_results_survived[model_name] = {'acc': result.holdout_acc, 'params' : param_comb[np.argmax(result.cv_test_acc)]}



    # LOGISTIC REGRESSION
    model_name = 'logistic_regression'
    param_comb = ParameterGrid({'C' :  [0.001,0.01,0.1,1,10,100,1000, 10000, 100000, 1000000], 
                                'solver' : ['newton-cg', 'lbfgs', 'sag', 'saga'], 
                                'multi_class' : ['ovr', 'multinomial']})

    result = crosval_train(model_name, X, Y, X_holdout, Y_holdout, param_comb, num_fold, metric, greater)
    results_to_plot[model_name] = result
    models_results_survived[model_name] = {'acc': result.holdout_acc, 'params' : param_comb[np.argmax(result.cv_test_acc)]}



    # KMEANS
    model_name = 'kmeans'
    dt = KMeans(n_clusters = 2, n_init=10, max_iter=300, random_state = 42)
    dt.fit(X, Y)
    holdout_acc = metrics.accuracy_score(Y_holdout, dt.predict(X_holdout))
    train_acc = metrics.accuracy_score(Y, dt.predict(X))

    models_results_survived[model_name] = {'acc': holdout_acc, 'params' : {} }
    results_to_plot[model_name] = ModelResults([train_acc], [train_acc], [holdout_acc])



    # NAIVE BAYES
    model_name = 'naive_bayes'
    bnb = BernoulliNB(binarize=0.0)
    bnb.fit(X,Y)
    holdout_acc = metrics.accuracy_score(Y_holdout, bnb.predict(X_holdout))
    train_acc = metrics.accuracy_score(Y, dt.predict(X))

    models_results_survived[model_name] = {'acc': holdout_acc, 'params' : {} }
    results_to_plot[model_name] = ModelResults([train_acc], [train_acc], [holdout_acc])



    # KNN
    # we need to remove nominal values first !
    deck_onehot = pd.get_dummies(df_for_survived.deck, prefix='deck')
    df_for_knn = pd.concat([df_for_survived.drop(columns=['deck', 'military', 'noble']), deck_onehot], axis=1)
    X_onehot, X_holdout_onehot, Y_onehot, Y_holdout_onehot = train_test_split(df_for_knn.drop(columns = ['survived']), df_for_knn['survived'], test_size=0.25, random_state=42)

    # knn
    model_name = 'knn'
    param_comb = ParameterGrid({'n_neighbors' : range(1,30), 'p' : [1,2,3,4,5]})

    result = crosval_train(model_name, X, Y, X_holdout, Y_holdout, param_comb, num_fold, metric, greater)
    results_to_plot[model_name] = result
    models_results_survived[model_name] = {'acc': result.holdout_acc, 'params' : param_comb[np.argmax(result.cv_test_acc)]}


    ###############################################################################################

    # Least but not last print best results
    print('{0:25}{1:23}{2}'.format('model', 'houldout_acc', 'best_hyperparams'))
    print()
    dictList = []

    for key, value in models_results_survived.items():
        temp = [key,value]
        dictList.append(temp)
    dictList.sort(key=lambda e: e[1]['acc'], reverse=True)
    for item in dictList:
        print('{0:25}{1}\t\t\t{2}'.format(item[0], item[1]['acc'], item[1]['params']))
    print()

    # And plot it all
    i = 1
    plt.figure(figsize=(25,20))       # set dimensions of the figure
    for model_name, result in results_to_plot.items():
        plt.subplot(4,2, i)         # create subplots on a grid with 3 rows and 2 columns
        plt.plot(result.cv_train_acc,'or-')
        plt.plot(result.cv_test_acc,'ob-')
        plt.plot([0,len(result.cv_test_acc)], [result.holdout_acc,result.holdout_acc],'g-')
        plt.xlabel('max_depth')
        plt.ylabel('accuracy')
        plt.legend(['train_acc_mean', 'test_acc_mean','houdlout_acc'])
        plt.title(model_name, fontsize=14, fontweight='bold')
        i +=1

    plt.show()
    best_model_survived = dictList[0][0]
    best_params_survived = models_results_survived[best_model_survived]['params']
```
